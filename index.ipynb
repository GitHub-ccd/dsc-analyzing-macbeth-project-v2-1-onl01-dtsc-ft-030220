## Getting the Data
Here we start by importing a Python package called `requests`. We'll use it to pull the transcript of Macbeth from the [Project Gutenberg](https://www.gutenberg.org/) website. We'll also preview a few details about what is now stored in the variable `macbeth`. As you can see, it's a string with 120,253 characters - the first 500 of which are printed below. 

```python
import requests
macbeth = requests.get('http://www.gutenberg.org/cache/epub/2264/pg2264.txt').text

print(type(macbeth), "\n#####################")
print(len(macbeth), "\n#####################")
print(macbeth[:100], "\n#####################")

# import re  # regex is useless here. macbeth string has special characters 
#print(re.search(r'pattern to(.*?)$', string).group(1))

start=macbeth.find('David Reed')
macbeth_clean=macbeth[start:]

print("\n\n##############Here is the clean version########\n\n", macbeth_clean[:100])
```

```python
import numpy as np
import matplotlib.pyplot as plt
import re
%matplotlib inline

# Your code here
'''
# result = re.sub(r'[^a-zA-Z0-9]', ' ',text) # substitute anything not alphanumeric to space 
#clean_macbeth= re.sub(r'[^a-zA-Z0-9]', ' ',macbeth)
clean_macbeth= re.sub(r'\*', ' ',macbeth)
none of this is needed if i remove the useless first few pages 
'''

macbeth_words= re.sub(r'[^a-zA-Z]', ' ',macbeth_clean)  # remove all numbers from clean string 
words=macbeth_words.split()
#print(len(words))
uniq_words=set(words)
#print(uniq_words)

word_stat={}
for word in uniq_words:
    word_stat[word]=0

for word in words:
    word_stat[word]+=1

print(len(word_stat))

import operator
sorted_d = sorted(word_stat.items(), key=operator.itemgetter(1))

sorted_word_stat={k: v for k, v in sorted(word_stat.items(), key=lambda item: item[1], reverse=True)} # works only in python 3.6+ 
sorted_word_stat



# Pseudo-code outline

# Split the transcript into words
# Create a dictionary
# Iterate through the text of Macbeth
# Update word counts
# Sort words by counts in descending order
# Create Bar Graph
# Include descriptive titles and labels
```

```python
sorted_word_stat
freq_words={k: v for i, (k, v) in enumerate(sorted_word_stat.items()) if i < 25}
freq_words
```

```python
# Import matplotlib
import matplotlib.pyplot as plt

# Set plot space as inline for inline plots and qt for external plots
%matplotlib inline

plt.figure(figsize=(10,6))

# Use the bar() function to create a plot using the above values of x and y. Add a label.
plt.bar(freq_words.keys(), freq_words.values())

plt.xticks(rotation='vertical')
plt.xlabel('word')
plt.ylabel('word count')

plt.title('Most frequent words in Macbeth')
#plt.legend()

# Output the final plot
plt.show()
```
